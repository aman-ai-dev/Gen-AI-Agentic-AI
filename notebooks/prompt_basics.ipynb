```python
from transformers import pipeline

llm = pipeline("text-generation", model="gpt2")

def run_prompt(system, task, constraints, input_text):
    prompt = f"""
SYSTEM:
{system}

TASK:
{task}

CONSTRAINTS:
{constraints}

INPUT:
{input_text}
"""
    return llm(prompt, max_length=120)[0]["generated_text"]
```

```python
output = run_prompt(
    system="You are a helpful AI assistant.",
    task="Explain what Agentic AI is.",
    constraints="Explain in simple words. Max 4 lines.",
    input_text=""
)

print(output)
```
